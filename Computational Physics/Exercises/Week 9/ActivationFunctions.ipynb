```python
import torch
import matplotlib.pyplot as plt
import numpy as np

from cpclasses.PlotTools import *
```

```python
activation_functions = [torch.nn.ReLU(), torch.nn.Sigmoid(), torch.nn.Softplus(), torch.nn.Softmax(dim=0), torch.nn.LeakyReLU(), torch.nn.SiLU()]
names = ['ReLU', 'Sigmoid', 'Softplus', 'Softmax', 'LeakyReLU', 'SiLU']

fig, axs = plt.subplots(2,3, figsize=(12,10))


for ax, activation, name in zip(axs.flatten(), activation_functions, names):
    x = torch.linspace(-10, 10, 100)
    y = activation(x)
    
    #ax.set_title(activation.__name__)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(name)
    coordinate_axes(ax)
    ax.plot(x, y)

save_plot(fig, 'activation_functions.png')
```
