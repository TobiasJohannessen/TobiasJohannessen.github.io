```python
f = lambda x: x - 0.5
def g(x):
    x0 = 1.25
    return torch.where(x>x0, f(x), f(x0))

x_train = torch.tensor([-0.5, 0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4., ])
y_train = g(x_train)
```

```python
activation_function = torch.relu

fig, axs = plt.subplots(2,2, figsize=(10,10))
for ax in axs.flatten():
    model = TwoLayerCustomActivation(activation_function)
    model.train_model(x_train, y_train, epochs=1000, lr=0.01)
    model.plot(ax, -2, 5)
    ax.plot(x_train, y_train, 'o', label='Training data')
    coordinate_axes(ax)
    format_axis(ax, legend=False)
    
fig.suptitle('Two layer network with ReLU', fontsize=20)
```

```python
activation_function = torch.nn.LeakyReLU()

fig, axs = plt.subplots(2,2, figsize=(10,10))
for ax in axs.flatten():
    model = TwoLayerCustomActivation(activation_function)
    model.train_model(x_train, y_train, epochs=1000, lr=0.01)
    model.plot(ax, -2, 5)
    ax.plot(x_train, y_train, 'o', label='Training data')
    coordinate_axes(ax)
    format_axis(ax, legend=False)
    
fig.suptitle('Two layer network with LeakyReLU', fontsize=20)
```

```python
activation_functions = [torch.relu, torch.nn.LeakyReLU(), torch.nn.SiLU(), torch.nn.Softplus()]
names = ['ReLU', 'LeakyReLU', 'SiLU', 'Softplus']


fig, axs = plt.subplots(4,2, figsize=(10,10), gridspec_kw={'hspace': 0.5})

for i, ax in enumerate(axs.flatten()):
    index = int(np.floor(i/2))
    activation_function = activation_functions[index]
    model = TwoLayerCustomActivation(activation_function)
    model.train_model(x_train, y_train, epochs=1000, lr=0.01)
    model.plot(ax, -2, 5, color = f'C{index}')
    ax.plot(x_train, y_train, 'ko', label='Training data')
    coordinate_axes(ax)
    format_axis(ax, legend=False, title=names[index])
```
