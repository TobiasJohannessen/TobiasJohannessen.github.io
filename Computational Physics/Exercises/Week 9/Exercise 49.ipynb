```python
f = lambda x: x - 0.5
def g(x):
    x0 = 1.25
    return torch.where(x>x0, f(x), f(x0))

x_train = torch.tensor([-0.5, 0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4., ])
y_train = g(x_train)

x_test = torch.linspace(-2, 5, 100)
```

```python
activation_function = torch.relu

fig, axs = plt.subplots(2,2, figsize=(10,10))
for ax in axs.flatten():
    model = TwoLayerCustomActivation(activation_function)
    model.fit(x_train, y_train, epochs=1000, lr=0.001)
    model.plot(ax, x_test, 1)
    ax.plot(x_train, y_train, 'o', label='Training data')
    coordinate_axes(ax)
    format_axis(ax, legend=False)
    
fig.suptitle('Two layer network with ReLU', fontsize=20)
```

```python
activation_function = torch.nn.LeakyReLU()

fig, axs = plt.subplots(2,2, figsize=(10,10))
for ax in axs.flatten():
    model = TwoLayerCustomActivation(activation_function)
    model.fit(x_train, y_train, epochs=2000, lr=0.001)
    model.plot(ax, x_test, 1)
    ax.plot(x_train, y_train, 'o', label='Training data')
    coordinate_axes(ax)
    format_axis(ax, legend=False)
    
fig.suptitle('Two layer network with LeakyReLU', fontsize=20)
```

```python
activation_function = torch.nn.SiLU()

fig, axs = plt.subplots(2,2, figsize=(10,10))
for ax in axs.flatten():
    model = TwoLayerCustomActivation(activation_function)
    model.fit(x_train, y_train, epochs=10000, lr=0.001)
    model.plot(ax, x_test, 1)
    ax.plot(x_train, y_train, 'o', label='Training data')
    coordinate_axes(ax)
    format_axis(ax, legend=False)
    
fig.suptitle('Two layer network with SiLU', fontsize=20)
```

```python
activation_function = torch.nn.Softplus()

fig, axs = plt.subplots(2,2, figsize=(10,10))
for ax in axs.flatten():
    model = TwoLayerCustomActivation(activation_function)
    model.fit(x_train, y_train, epochs=1000, lr=0.01)
    model.plot(ax, x_test, 1)
    ax.plot(x_train, y_train, 'o', label='Training data')
    coordinate_axes(ax)
    format_axis(ax, legend=False)
    
fig.suptitle('Two layer network with SoftPlus', fontsize=20)
```

```python
activation_functions = [torch.relu, torch.nn.LeakyReLU(), torch.nn.SiLU(), torch.nn.Softplus()]
names = ['ReLU', 'LeakyReLU', 'SiLU', 'Softplus']


x_train = torch.tensor([-0.5, 1, 2, 2.5, 4., ])
x_test = torch.linspace(-2, 4, 2)
y_train = g(x_train)

fig, axs = plt.subplots(4,2, figsize=(8,12), gridspec_kw={'hspace': 0.05})

for i, ax in enumerate(axs.flatten()):
    index = int(np.floor(i/2))
    activation_function = activation_functions[index]
    model = Two_8_LayerCustomActivation(activation_function)
    model.fit(x_train, y_train, epochs=1000, lr=0.01)
    model.plot(ax, x_test, 1, color = f'C{index}')
    ax.plot(x_train, y_train, 'ko', label='Training data')
    coordinate_axes(ax)
    format_axis(ax, legend=False, title=names[index])
```
