```python
pos = [[0,0], [1,0], [0.5, 1], [1.5, 1]]

PS = PeriodicSystem(calc = PeriodicLennardJones(eps=5), pos = pos, box = 2*np.eye(2))


fig, axs = plt.subplots(1,3, figsize = (24,8))

ax = axs[0]
PS.draw_periodic(ax, size = 250)


ax.set(xlim = (-10,10), ylim = (-10,10))
ax.set_title(r'$E_{pot} = $' + f'{PS.potential_energy:.2f}')



pos = np.array([[0,0], [1,0], [0.5, 1], [1.5, 1]])

pos_4 = pos.copy()

pos_4 = np.concatenate([pos_4, pos + [2,0], pos + [0,2], pos + [2,2]])

PS = PeriodicSystem(calc = PeriodicLennardJones(eps=5), pos = pos_4, box = 4*np.eye(2))

ax = axs[1]

PS.draw_periodic(ax, size = 250)


ax.set(xlim = (-10,10), ylim = (-10,10))
ax.set_title(r'$E_{pot} = $' + f'{PS.potential_energy:.2f}')


pos_4_4 = pos_4.copy()

pos_4_4 = np.concatenate([pos_4_4, pos_4 + [4,0], pos_4 + [0,4], pos_4 + [4,4]])

PS = PeriodicSystem(calc = PeriodicLennardJones(eps=5), pos = pos_4_4, box = 8*np.eye(2))

ax = axs[2]

PS.draw_periodic(ax, size = 250)


ax.set(xlim = (-10,10), ylim = (-10,10))
ax.set_title(r'$E_{pot} = $' + f'{PS.potential_energy:.2f}')

save_plot(fig, 'PeriodicSystem, growing.png')
```
