```python
import numpy as np
import matplotlib.pyplot as plt
import Clustering
from importlib import reload
reload(Clustering)
from Clustering import *
```

```python
pos_flat = np.loadtxt('lj10clusters.txt')
positions = pos_flat.reshape(-1,pos_flat.shape[1]//2,2)
```

```python


N_clusters = 4

cluster_rows = int(np.ceil(N_clusters))
cluster_cols = 2
fig, axs = plt.subplots(cluster_rows, cluster_cols, figsize=(10, N_clusters * 5))

clusters = [AtomicCluster(pos, descriptor_method=DistanceMoments) for pos in positions[:N_clusters]]

for ax, cluster in zip(axs, clusters):
    print(cluster.cluster)
    cluster.draw(ax[0])
    ax[0].set_aspect('equal')
    ax[0].set(xlim=(-4, 4), ylim=(-4, 4))
    ax[0].grid()
    cluster.draw_descriptor(ax[1])
    ax[1].set_aspect('equal')
    
print(cluster.descriptor_method)
```

```python
fig, ax = plt.subplots()

clusters = [AtomicCluster(pos, descriptor_method=DistanceMoments) for pos in positions]

for cluster in clusters:
    cluster.draw_cluster(ax)
ax.grid()

colors = [f'C{i}' for i in range(1,7)]
labels = [f'Cluster {i}' for i in range(1,7)]
ax.legend( handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=c, markersize=10, alpha = 0.5) for c in colors], labels=labels, loc='upper left')
```
