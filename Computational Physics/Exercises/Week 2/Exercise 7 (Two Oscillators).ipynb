```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from scipy.integrate import quad
%matplotlib inline
from importlib import reload
import Thermodynamics
reload(Thermodynamics)
from Thermodynamics import *
```

```python
harmonic_oscillator1 = HarmonicOscillator(k = 1, x_0 = 0, x_range = (-3, 3), N_bins=50, kT=0.15)
harmonic_oscillator2 = HarmonicOscillator(k = 10, x_0 = 1, x_range = (-3, 3), N_bins=50, kT = 0.15)
fig, ax = plt.subplots()

ax.set(xlim=(-3, 3), ylim=(0, 5))
i = 1
for harmonic_oscillator in [harmonic_oscillator1, harmonic_oscillator2]:
    print(f'Harmonic Oscillator {i}\n')
    i += 1
    harmonic_oscillator.plot(ax)
    print(f'Direct integration gives {harmonic_oscillator.direct_integration():.2g} kT')
    print(f'MC sampling gives {harmonic_oscillator.V_avg_MC:.2g} kT')
    print('\n')


print(f'Non-Boltzmann sampling (1->2) gives <V> = {harmonic_oscillator1.non_boltzmann_sample(harmonic_oscillator2):.3g} kT')
print(f'Non-Boltzmann sampling (2->1) gives <V> = {harmonic_oscillator2.non_boltzmann_sample(harmonic_oscillator1):.3g} kT')

ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

ax.set_title('Two Harmonic Oscillators', fontsize = 18)
ax.set(xlabel = 'x', ylabel = 'V(x)')

plt.savefig('Figures/Two_Harmonic_Oscillators.png', bbox_inches='tight')
```
