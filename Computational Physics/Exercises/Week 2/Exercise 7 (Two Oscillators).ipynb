```python
from Thermodynamics import *
```

```python
HO1 = HarmonicOscillator(kT = 0.15, k = 1, x_0 =0, x_range=[-3,3])
HO2 = HarmonicOscillator(kT = 0.15, k = 10, x_0 =1, x_range=[-1, 3])
```

```python
fig, ax = plt.subplots()

HO1.plot_V(ax)
HO1.plot_P(ax)
HO2.plot_V(ax, color='tab:orange')
HO2.plot_P(ax, color='tab:orange')

ax.set(xlabel='x', ylabel='V(x), P(x)', xlim=(-3, 3), ylim=(0, 5))
```

```python
print(f'Harmonic Oscillator 1: <V> = {HO1.V_avg_direct:.3f}')
print(f'Harmonic Oscillator 1: <V> = {HO1.V_avg_MC:.3f}')
print(f'Harmonic Oscillator 2: <V> = {HO2.V_avg_direct:.3f}')
HO2.sample_n_MC(1000000)
print(f'Harmonic Oscillator 2: <V> = {HO2.V_avg_MC:.3f}')

```
