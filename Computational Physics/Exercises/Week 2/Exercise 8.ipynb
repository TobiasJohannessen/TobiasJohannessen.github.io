```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from scipy.integrate import quad
%matplotlib inline
from importlib import reload
import Thermodynamics
reload(Thermodynamics)
from Thermodynamics import *

from cpclasses.PlotTools import *
```

```python
harmonic_potential = lambda x: 0.5 * 1 * x**2

h_o_rw = Potential(harmonic_potential, kT = 0.15, x_range = [-2, 2], N_bins = 100, type = 'Harmonic', method = 'Random Walk')

Ns_for_MCMC = np.logspace(2, 5, 4)
fig, axs = plt.subplots(2,2, figsize=(10,10))


for N, ax in zip(Ns_for_MCMC, axs.flatten()):
    ax.set(xlim=(-2, 2))
    h_o_rw.plot_mcmc(ax, N = int(N))
    h_o_rw.plot_V(ax)
    ax.set_title('N = {}'.format(int(N)))

save_plot(fig, 'Harmonic_RW, Varying N.png')
```
