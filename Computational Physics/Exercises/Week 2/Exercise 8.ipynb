```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from scipy.integrate import quad
%matplotlib inline
from importlib import reload
import Thermodynamics
reload(Thermodynamics)
from Thermodynamics import *
```

```python
harmonic_potential = lambda x: 0.5 * 1 * x**2

h_o_rw = Potential(harmonic_potential, kT = 0.15, x_range = [-2, 2], N_bins = 40, type = 'Harmonic')

Ns_for_MCMC = np.logspace(2, 5, 4)
fig, axs = plt.subplots(1,4, figsize=(20,5))


for N, ax in zip(Ns_for_MCMC, axs):
    ax.set(xlim=(-2, 2))
    h_o_rw.plot_mcmc(ax, N = int(N), method = 'Random Walk')
    h_o_rw.plot_V(ax)
    ax.set_title('N = {}'.format(int(N)))
```
