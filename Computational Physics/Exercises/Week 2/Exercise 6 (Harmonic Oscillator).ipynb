```python
harmonic_oscillators = [HarmonicOscillator(k = 1, x_0 = 0), HarmonicOscillator(k = 2, x_0 = 0), HarmonicOscillator(k = 0.5, x_0 = 0, x_range=(-4,4))]

fig, axs = plt.subplots(1,len(harmonic_oscillators), figsize = (15,8))



for harmonic_oscillator, ax in zip(harmonic_oscillators, axs):
    print(f'k = {harmonic_oscillator.k}')
    
    harmonic_oscillator.thermodynamics()
    ax.set_title(f'Direct integration gives <V> = {harmonic_oscillator.direct_integration():.3g}\n' 
                 + f'Numerical Derivative gives C_V = {harmonic_oscillator.numerical_derivative(dT = 0.0001):.3g}\n' 
                 + f'Fluctuations gives C_V = {harmonic_oscillator.fluctuations():.3g}\n' 
                 + f'Monte Carlo gives <V> = {harmonic_oscillator.direct_MC(N = 100000):.3g} \n'
                 + f'k = {harmonic_oscillator.k}')
    harmonic_oscillator.plot_V(ax)
    harmonic_oscillator.plot_P(ax, color='tab:orange')
    harmonic_oscillator.plot_mcmc(ax, color='tab:green')
    ax.legend()
    ax.set(xlim=(-2,2), ylim=(0,4))
    ax.set_aspect('equal')

ax.set(xlim=(-4,4), ylim=(0,8))
ax.set_aspect('equal')



#save_plot('Figures\\3_Harmonic_Oscillators,2.png', bbox_inches='tight')
```
