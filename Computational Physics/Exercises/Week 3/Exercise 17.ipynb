```python
fig, axs = plt.subplots(2, 4, figsize = (16, 8))
LJ = LennardJones()
for ax in axs.flatten():
    ax.set(xlim = (-5,5), ylim = (-5,5))
    cluster = AtomicCluster(LJ, N = 5)
    cluster.pos[0] = np.array([0, 0])
    cluster.static[0] = True
    #cluster.draw(ax, radius = 33, alpha = 0.4, color ='white', edge = True, center=True)
    basin_hopping(cluster, N = 100)
    cluster.draw(ax, radius = 33, force_draw = True, alpha  = 1, center = True)
    ax.grid()
    ax.title.set_fontsize(18)
```

```python
save_plot(fig, 'basin_hopping, 5 atoms')
```
