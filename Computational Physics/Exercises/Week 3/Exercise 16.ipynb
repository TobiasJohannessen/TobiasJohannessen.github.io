```python
from cpclasses import *
```

```python
LJ = LennardJones()
static_points = np.array([[-1, 0], [1, 0]])
static = [True, True]
cluster = AtomicCluster(LJ, pos = static_points, static = static)

x = np.linspace(-4,4,400)
y = np.linspace(-4,4,400)

Xs, Ys = np.meshgrid(x,y)

Zs = np.zeros_like(Xs)
for i in range(len(x)):
    for j in range(len(y)):
        Zs[i,j] = LJ.energy(np.array([[Xs[i,j], Ys[i,j]], static_points[0], static_points[1]]))
        if Zs[i,j] > 0:
            Zs[i,j] = 0
```

```python

particles = ParticleSwarm(cluster, 10, ((-4, 4), (-4, 4)), w=0.8, c1=0.2, c2=0.5)

fig, ax = plt.subplots()

contour = ax.contourf(Xs,Ys,Zs, levels = 20, cmap = 'hot')
plt.colorbar(contour)

particles.plot_particles(ax)


particles.evaluate(particles.positions[0])
```
