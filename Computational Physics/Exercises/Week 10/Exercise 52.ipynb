```python

from cpclasses import *
import torch

```

```python
w = 1
b = 0.5
xs_train = torch.rand(32) * 6 - 2
ys_train = w * xs_train + b + 0.5 * torch.randn(32)
```

```python
linearModel = CustomDeepNetwork(n_hidden=0, n_neurons=1, activation='relu')
NonLinearModel = CustomDeepNetwork(n_hidden=3, n_neurons=8, activation='relu')
```

```python
linearModel.fit(xs_train, ys_train, epochs=10000, lr=0.01)
NonLinearModel.fit(xs_train, ys_train, epochs=10000, lr=0.01)
```

```python
fig, axs = plt.subplots(1,2)



linearModel.plot(axs[0], xs_train, ys_train)
axs[0].plot(xs_train, ys_train, 'ro', label='')
axs[0].set_title('Linear Model')

NonLinearModel.plot(axs[1], xs_train, ys_train)
axs[1].plot(xs_train, ys_train, 'ro', label='')
axs[1].set_title(r'NN$_{(1 \times 8\times 8 \times 8 \times 1)}$')  

for ax in axs:
    ax.grid()
    ax.set(xlabel='x', ylabel='y', aspect='equal')
    ax.legend()
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,15))
decays = [0, 1e-5, 1e-3, 1e-1]

runs = 4

for ax, decay in zip(axs.flatten(), decays):
    for i in range(runs):
        NonLinearModel = CustomDeepNetwork(n_hidden=3, n_neurons=8, activation='silu')
        NonLinearModel.fit(xs_train, ys_train, epochs=5000, lr=0.1, weight_decay=decay)
        NonLinearModel.plot(ax, xs_train, ys_train, label = '', color = f'C{i}')
    ax.plot(xs_train, ys_train, 'o')
    ax.set_title(f'Weight Decay: {decay:.0g}')
    ax.grid()
    ax.set(xlabel='x', ylabel='y', aspect='equal')
    ax.legend()

save_plot(fig, 'weight_decay, silu')
```
